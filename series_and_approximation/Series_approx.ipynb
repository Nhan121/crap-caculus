## Series vs convergence of series
For example,

$$ s_n = \sum_{k=0}^{n} \dfrac{1}{(k+1)^2} $$

```python
def your_series(n):     
    s = 0
    for k in range(n+1):
        s += 1 / (k+1)**2    
    return s
dvn_aps.series_n_epsilon(n_min=1, n_max=500, n_steps=50, 
                 your_series = your_series, series_limit = np.pi**2/6)
```

- Now, we have known that

$$ e^{-1} = \sum_{n=0}^{\infty} \dfrac{(-1)^n}{n!} $$

or

$$ \lim_{n \to \infty} s_n = e^{-1} $$

where

$$ s_n = \sum_{k=0}^{n} \dfrac{(-1)^k}{k!} $$

```python
def your_series(n):     
    s = 0
    for k in range(0, n+1):
        s += (-1)**k / np.math.factorial(k)    
    return s
dvn_aps.series_n_epsilon(n_min=1, n_max=12, n_steps=1, 
                 your_series = your_series, series_limit = np.exp(-1))
```
