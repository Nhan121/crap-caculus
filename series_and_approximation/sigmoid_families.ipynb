```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style = "darkgrid", 
                  rc = {"axes.spines.right": False, "axes.spines.top": False})

x = np.linspace(-6,6,100)
dvn_aps.depth_view_actv_func(np.tanh, x, n_list = [1,5,10,15,25,125], func_name = 'tanh')
plt.show()
```

```python
def sigmoid(x): return 1 / (1 + np.exp(-x))

def n_layers_fx(fx, x, n):
    
    y = fx(x)
    if n == 1:
        return fx(x)
    else:
        for k in range(n):
            y = fx(y)
    return y

fig, ax = plt.subplots(1, 1, figsize = (20, 8), dpi = 100)
x = np.linspace(-4, 4, 500)
for n in [1,2,3]:
    y = [n_layers_fx(sigmoid, t, n = n) for t in x]
    ax.plot(x, y, label = 'n_layers={}'.format(n))
    ax.grid('violet')

plt.title("$f_1(x) = sigmoid(x)$\n $f_2(x) = sigmoid(sigmoid(x))$\n ...\n $f_n(x) = sigmoid(sigmoid(\ldots(x)))$\n where n is the number of layers")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
def fx(x): return x / (1 + abs(x))

fig, ax = plt.subplots(1, 1, figsize = (20, 8), dpi = 100)
x = np.linspace(-4, 4, 500)
for n in [1,2,3,9,81]:
    y = [n_layers_fx(fx, t, n = n) for t in x]
    ax.plot(x, y, label = 'n_layers={}'.format(n))
    ax.grid(color = 'cyan', linestyle='--', linewidth = 1)

plt.title("$f_1(x) = \dfrac{x}{1+|x|} $\n $f_2(x) = \dfrac{f_1x}{1+|f_1x|}$\n ...\n $f_n(x) = \dfrac{f_{n-1}x}{1+|f_{n-1}x|}$\n where n is the number of layers")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
def fx(x): return x / np.sqrt(1 + x*x)

fig, ax = plt.subplots(1, 1, figsize = (20, 8), dpi = 100)
x = np.linspace(-4, 4, 500)
for n in [1,2,3,9,81]:
    y = [n_layers_fx(fx, t, n = n) for t in x]
    ax.plot(x, y, label = 'n_layers={}'.format(n))
    ax.grid(color = 'cyan', linestyle='--', linewidth = 1)

plt.title("$f_1(x) = \dfrac{x}{\sqrt{x^2+1}} $\n $f_2(x) = \dfrac{f_1(x)}{\sqrt{(f_1(x))^2+1}} $\n ...\n $f_n(x) = \dfrac{f_{n-1}(x)}{\sqrt{(f_{n-1}(x))^2+1}}$\n where n is the number of layers")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
a = (2/np.pi)
def fx(x): return a*np.arctan(x/a)

fig, ax = plt.subplots(1, 1, figsize = (20, 8), dpi = 100)
x = np.linspace(-6, 6, 500)
for n in [1,2,3,9,81]:
    y = [n_layers_fx(fx, t, n = n) for t in x]
    ax.plot(x, y, label = 'n_layers={}'.format(n))
    ax.grid(color = 'cyan', linestyle='--', linewidth = 1)

plt.title("$f_1(x) = c*arctan(x / c) $\n $f_2(x) =c*arctan(f_1(x)/c) $\n ...\n $f_n(x) = c*arctan(f_{n-1}(x)/c)$\n where n is the number of layers and $c=\dfrac{2}{\pi}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
